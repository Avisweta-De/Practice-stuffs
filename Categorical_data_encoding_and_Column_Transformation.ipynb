```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
df2= pd.read_csv('covid_toy.csv')
df2.head()
```

```python
df2.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X2_train,X2_test,y2_train,y2_test = train_test_split(df2.drop(columns=['has_covid']),df2['has_covid'],
                                                test_size=0.2)
```

```python
X2_train
```

```python
si = SimpleImputer()
X2_train_fever = si.fit_transform(X2_train[['fever']])

# also the test data
X2_test_fever = si.fit_transform(X2_test[['fever']])

X2_train_fever.shape

```

```python
# Ordinalencoding -> cough
oe = OrdinalEncoder(categories=[['Mild','Strong']])
X2_train_cough = oe.fit_transform(X2_train[['cough']])

# also the test data
X2_test_cough = oe.fit_transform(X2_test[['cough']])

X2_train_cough.shape
```

```python
# OneHotEncoding -> gender,city
ohe = OneHotEncoder(drop='first', sparse_output = False)
X2_train_gender_city = ohe.fit_transform(X2_train[['gender','city']])

# also the test data
X2_test_gender_city = ohe.fit_transform(X2_test[['gender','city']])

X2_train_gender_city.shape
```

```python

# Extracting Age
X2_train_age = X2_train.drop(columns=['gender','fever','cough','city']).values

# also the test data
X2_test_age = X2_test.drop(columns=['gender','fever','cough','city']).values

X2_train_age.shape
```

```python
X2_train_transformed = np.concatenate((X2_train_age,X2_train_fever,X2_train_gender_city,X2_train_cough),axis=1)
# also the test data
X2_test_transformed = np.concatenate((X2_test_age,X2_test_fever,X2_test_gender_city,X2_test_cough),axis=1)
print("Shape of transformed training data:")
print(X2_train_transformed.shape)
```

```python
#importance of column transformation
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])],remainder='passthrough')

transformer.transform(X2_test).shape
```
